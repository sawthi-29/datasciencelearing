```python
import pandas as pd
```

```python
df = pd.read_csv('/content/employee_attendance (1).csv')
```

```python
df
```

```python
df["login_date"] = pd.to_datetime(df["login_date"])
df["login_time"] = pd.to_datetime(df["login_time"])
df.info
```

```python
df["year"] = df["login_date"].dt.year
df["month"] = df["login_date"].dt.month
df["day"] = df["login_date"].dt.day
df.info
```

```python
df["month_name"] = df["login_date"].dt.month_name()
df["dat_name"] = df["login_date"].dt.day_name()
df.info
```

```python
df['hour'] = df['login_time'].dt.hour
df['minute'] = df['login_time'].dt.minute
df['second'] = df['login_time'].dt.second
df.info
```
