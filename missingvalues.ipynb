```python
import seaborn as sns
df = sns.load_dataset('titanic')
df
```

```python
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df["age"]
```

```python
df["Age_Mean"] = df["age"].fillna(df["age"].mean())
print(df["Age_Mean"])
```

```python
df[["Age_Mean","age"]]
```

```python
df["Age_Median"] = df["age"].fillna(df["age"].median())
print(df["Age_Median"])
```
