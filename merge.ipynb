```python
#concat in pandas
concat() is used when you want to combine mutliple dataframes
```

```python
import pandas as pd
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
print(india_weather)
```

```python
paris_weather = pd.DataFrame({
    "city": ["paris","london","newyork"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
print(paris_weather)
```

```python
df = pd.concat([india_weather,paris_weather],ignore_index=True)
df
```
